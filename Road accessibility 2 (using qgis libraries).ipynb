## Load a shapefile and plot it

Load a shapefile with some roads.

See http://docs.qgis.org/testing/en/docs/pyqgis_developer_cookbook/loadlayer.html.

```python
current_dir = %pwd
path = os.path.join(current_dir, "data", "sample_roads.shp")
roads_layer = QgsVectorLayer(path, "roads", "ogr")
if roads_layer.isValid():
    print("loaded %s, %i features" % (path, roads_layer.featureCount()))
else:
    print("*** failed to load %s" % (path))
```

Plot the layer to check it loaded and looks right.

```python
index_objectid = roads_layer.fieldNameIndex("OBJECTID")
index_registered = roads_layer.fieldNameIndex("REGISTERED")
index_streettype = roads_layer.fieldNameIndex("STREETTYPE")

fig, ax = plt.subplots(figsize=(10, 10))
for road in roads_layer.getFeatures():
    geom = road.geometry()
    if geom.type() != QGis.Line:
        print("*** feature is not a line")
    x, y = zip(*geom.asPolyline())
    if road.attributes()[index_streettype] == "AVE":
        ax.plot(x, y, "r-")
    else:
        ax.plot(x, y, "k-")
ax.set_aspect("equal")
ax.grid()
```

## Convert the layer to a graph

Build a graph of the layer using QGIS network analysis library.

See http://docs.qgis.org/testing/en/docs/pyqgis_developer_cookbook/network_analysis.html.

```python
road = roads_layer.getFeatures().next()
for field in roads_layer.pendingFields():
    print("%s %s" % (field.name(), field.typeName()))
```

```python
# Convert km/h to metres/second
def kmh_to_mps(kmh):
    return kmh * 1000.0 / 3600
kmh_to_mps(36)
```

```python
# Based on http://gis.stackexchange.com/questions/94457/how-can-i-use-python-qgis-network-analysis-library-to-calculate-short-path-based
speed_default = kmh_to_mps(10)
speed_avenue = kmh_to_mps(50)
class RoadTravelTimeProperter(QgsArcProperter):
    def property(self, distance, feature):
        attrs = feature.attributes()
        print("computing travel time for objectid %i, %s %s, distance %s" %
              (attrs[index_objectid], attrs[index_registered], attrs[index_streettype], distance))
        street_type = attrs[index_streettype]
        val = distance / speed_default
        if (street_type == "AVE"):  # example: make Avenues faster
            val = distance / speed_avenue
        return val  # in seconds
    def requiredAttributes(self):
        return [index_objectid, index_registered, index_streettype]
road = roads_layer.getFeatures().next()
properter = RoadTravelTimeProperter()
properter.property(100, road)
```

```python
director = QgsLineVectorLayerDirector(roads_layer, -1, '', '', '', 3)
#properter = QgsDistanceArcProperter()
properter = RoadTravelTimeProperter()
director.addProperter(properter)
#crs = QgsCoordinateReferenceSystem(4326, QgsCoordinateReferenceSystem.PostgisCrsId)
crs = QgsCoordinateReferenceSystem(32617, QgsCoordinateReferenceSystem.PostgisCrsId)
#crs = QgsCoordinateReferenceSystem(102285, QgsCoordinateReferenceSystem.PostgisCrsId)
print "Description:", crs.description()
print "Is geographic:", crs.geographicFlag()
print "Map units:", crs.mapUnits()
print "Ellipsoid Acronym:", crs.ellipsoidAcronym()
builder = QgsGraphBuilder(crs, True, 0, crs.ellipsoidAcronym())
director.makeGraph(builder, [])
graph = builder.graph()
print("built graph, %i arcs, %i vertices" % (graph.arcCount(), graph.vertexCount()))
```

Plot the graph vertices to check it looks right.

```python
fig, ax = plt.subplots(figsize=(10, 10))
x = []
y = []
for vertex in range(0, graph.vertexCount()):
    x.append(graph.vertex(vertex).point()[0])
    y.append(graph.vertex(vertex).point()[1])
ax.plot(x, y, 'k.')
ax.set_aspect("equal")
ax.grid()
```

Plot the graph arcs to check how they look.

```python
fig, ax = plt.subplots(figsize=(10, 10))
n = 0
for arc in range(0, graph.arcCount()):
    v0 = graph.vertex(graph.arc(arc).inVertex())
    v1 = graph.vertex(graph.arc(arc).outVertex())
    x = [v0.point()[0], v1.point()[0]]
    y = [v0.point()[1], v1.point()[1]]
    ax.plot(x, y, 'ko-')
    n = n + 1
ax.set_aspect("equal")
ax.grid()
print("%i arcs plotted" % (n))
```

## Analyze the graph

Apply Dijkstra's algorithm to compute shortest paths from a random vertex.

```python
start_point_index = 50
start_point = graph.vertex(start_point_index)
print("start point %i %i" % (start_point.point()[0], start_point.point()[1]))
(tree, cost) = QgsGraphAnalyzer.dijkstra(graph, start_point_index, 0)
print("tree length %i, cost length %i" % (len(tree), len(cost)))
```

Show the shortest paths tree.

Labels show travel time in seconds from the start point (red) to each point.

```python
fig, ax = plt.subplots(figsize=(20, 20))
n = 0
# Plot all roads in grey underneath
for arc in range(0, graph.arcCount()):
    v0 = graph.vertex(graph.arc(arc).inVertex())
    v1 = graph.vertex(graph.arc(arc).outVertex())
    x = [v0.point()[0], v1.point()[0]]
    y = [v0.point()[1], v1.point()[1]]
    ax.plot(x, y, color="0.8")
# Plot the shortest path tree on top
for v in range(0, graph.vertexCount()):
    incoming_edge = tree[v]
    v0 = graph.vertex(v)
    if incoming_edge == -1:
        # no incoming edge to this vertex
        ax.plot([v0.point()[0]], [v0.point()[1]], "rh")
    else:
        arc = graph.arc(incoming_edge)
        v1 = graph.vertex(arc.outVertex())
        p0 = v0.point()
        p1 = v1.point()
        ax.plot([p0[0], p1[0]], [p0[1], p1[1]], "k-")
        ax.arrow(p1[0], p1[1], (p0[0] - p1[0]) * 0.5, (p0[1] - p1[1]) * 0.5, head_width=15, head_length=30, color="green")
        ax.text(p0[0], p0[1], str(int(cost[v])), horizontalalignment="center", verticalalignment="center",
                color="blue", path_effects=[matplotlib.patheffects.withStroke(linewidth=3, foreground="w")])
        n = n + 1
ax.set_aspect("equal")
ax.grid()
print("%i arrows plotted" % (n))
```

```python
x = []
y = []
z = []
for v in range(0, graph.vertexCount()):
    p = graph.vertex(v).point()
    x.append(p[0])
    y.append(p[1])
    z.append(cost[v])
fig, ax = plt.subplots(figsize=(20, 20))
xi = np.linspace(min(x), max(x), 100)
yi = np.linspace(min(y), max(y), 100)
zi = matplotlib.mlab.griddata(x, y, z, xi, yi, interp="linear")
plt.contourf(xi, yi, zi, range(0, 1200, 60))
cs = plt.contour(xi, yi, zi, range(0, 1200, 60), colors="#ffffff")
plt.clabel(cs, inline=1)
# plot roads on top
for arc in range(0, graph.arcCount()):
    v0 = graph.vertex(graph.arc(arc).inVertex())
    v1 = graph.vertex(graph.arc(arc).outVertex())
    x = [v0.point()[0], v1.point()[0]]
    y = [v0.point()[1], v1.point()[1]]
    ax.plot(x, y, "k")
ax.set_aspect("equal")
ax.grid()
```
